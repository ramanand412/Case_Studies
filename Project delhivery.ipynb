# Hypothesis testing

# Hypothesis testing between start scan to end scan and total time difference

```python
# null hypothesis h0 = there is no difference between the column start_scan_to_end_scan and total_min_diff
# alternate hypothesis h1 = there is difference between the column start_scan_to_end_scan and total_min_diff
# here two tail test is used to check the  null hypothesis true or not
```

```python
x=df.groupby(["trip_uuid","source_center"]).agg({"start_scan_to_end_scan":"max","total_time_diff":"max"})
x=x.groupby("trip_uuid").agg({"start_scan_to_end_scan":"sum","total_time_diff":"sum"}).reset_index()
x
```

```python
# concatenate two columns
p=np.concatenate((df["start_scan_to_end_scan"],df["total_time_diff"]))
```

```python
# take test sample c1 and c2 to corresponding columns and take mean of it 
# find the difference between means
# set a distribution of mean difference and append it to a list test sample
test_sample=[]
it=10000
for i in range(it):
    np.random.shuffle(p)
    c1=p[:144867]
    c2=p[144867:]
    mean_diff=np.mean(c1)-np.mean(c2)
    test_sample.append(mean_diff)
```

```python
# finding observed mean
observed_mean_diff=x["start_scan_to_end_scan"].mean()-x["total_time_diff"].mean()
observed_mean_diff
```

```python
# distribution of mean difference
sns.kdeplot(test_sample)
```

```python
# set a significance_level=0.05
# check the observed mean  between 2.5 percentile and 97.5 percentile of test sample
# import the library to find the percentile
from scipy.stats import norm
```

```python
#x=z*sigma+mu
#ucv is upper critical value above observed mean value means null hypothesis rejected
#lcv is lower critical value below observed mean value means null hypothesis rejected
ucv=norm.ppf(.975)*np.std(test_sample)+np.mean(test_sample)
lcv=norm.ppf(.025)*np.std(test_sample)+np.mean(test_sample)
print("lower critical value",lcv)
print("upper critical value",ucv)
```

```python
print("observed mean difference",observed_mean_diff)
```

#### observed values is inside of lcv and ucv so null hypothesis accepts so there is no difference between  total time taken to delivery and time difference of scaning before delivery and scaning after delivery

# Finding outliers of each numerical columns

```python
outlier_df=df.copy()
attrs=["start_scan_to_end_scan", 'actual_distance_to_destination',
       'actual_time', 'osrm_time', 'osrm_distance', 'segment_actual_time',
       'segment_osrm_time', 'segment_osrm_distance']
for i in attrs:
    q3=np.percentile(df[i],75)
    q1=np.percentile(df[i],25)
    q2=np.percentile(df[i],50)
    IQR=q3-q1
    upper=q3+1.5*IQR
    lower=q1-1.5*IQR
    print(f"lower limit of outliers for {i} are {lower}")
    print(f"upper limit of outlier for {i} are {upper}")
    print("----------------------------------------------------")
    outlier_df=outlier_df[~((outlier_df[i]<lower)|(outlier_df[i]>upper))]
```
