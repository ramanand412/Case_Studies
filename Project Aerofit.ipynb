### probability of male and female using each product

```python
x=df.groupby(["Product","Gender"])["Product"].count().to_frame("count").reset_index()
x=x.pivot("Product","Gender","count").reset_index()
x["sum"]=x["Female"]+x["Male"]
x["P(Female|Product)%"]=(x["Female"]/x["sum"]*100).round(2)
x["P(male|Product)%"]=(x["Male"]/x["sum"]*100).round(2)
x
```

- probability of KP281 purchased by male and female is equal ie 50%
- probability of KP481 purchased by male is 52% and female is 48%
- probability of KP781 purchased by male is bigger than female

### Convert Age columns to Bins

```python
df_new=df.copy()
labels1=["15-25","25-35","35-45","45-55"]
bins1=[15,25,35,45,55]
df_new["Age"]=pd.cut(df_new["Age"],bins=bins1,labels=labels1)
```

### Convert Income column to Bins

```python
labels1=["20k-40k","40k-60k","60k-80k","more_than_80k"]
bins1=[20000,40000,60000,80000,150000]
df_new["Income"]=pd.cut(df_new["Income"],bins=bins1,labels=labels1)
```

### Convert Miles to Bins

```python
labels1=["50-100","100-150","150-200","200-250",">250"]
bins1=[40,100,150,200,250,400]
df_new["Miles"]=pd.cut(df_new["Miles"],bins=bins1,labels=labels1)
df_new
```

## Marginal Probability of the purchase on the basis of income

```python
dd=df_new.groupby("Product")["Income"].value_counts().to_frame("count").reset_index()
dd.columns=["Product","Income_range","no_of_product_purchased"]

dd["P(Product purchased in an Income_range)"]=(dd["no_of_product_purchased"]/dd["no_of_product_purchased"].sum()*100).round(2)
dd
```

```python
plt.figure(figsize=(12,8))
df_new.groupby("Product")["Income"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

- Income range between 40k-60k have most probability ie 28% probability to purchase KP281
- Income range between >80k have least probability ie 0% probability to purchase KP281
- Income range between 40k-60k have most probability ie 24% probability to purchase KP481
- Income range between >80k have least probability ie 0% probability to purchase KP481
- Income range between >80k have most probability ie 10.6% probability to purchase KP781
- Income range between 20k-40k have least probability ie 0% probability to purchase KP781


## Marginal Probability of purchase on the basis of Age

```python
dd=df_new.groupby("Product")["Age"].value_counts().to_frame("count").reset_index()
dd.columns=["Product","Age","no_of_product_purchased"]
dd["P(Product_purchased in certain Age)"]=(dd["no_of_product_purchased"]/dd["no_of_product_purchased"].sum()*100).round(2)
dd
```

```python
plt.figure(figsize=(12,8))
df_new.groupby("Product")["Age"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

- age between 15-25 have most probability ie 18.9% probability to purchase KP281

- age between 15-25 have most probability ie 15.6% probability to purchase KP481
- age between 45-55 have least probability ie 0.57% probability to purchase KP481
- age between 15-25 & 25-35 have most probability ie 9.44% probability to purchase KP781
- age between 45-55 have least probability ie 1.11% probability to purchase KP781


## Conditional probability of product purchased in certain Income range

```python
x=pd.crosstab(df_new.Income,df_new.Product,margins=True).reset_index()
x
```

```python
x["P(KP281|income_range)"]=(x["KP281"]/x["All"]*100).round(2)
x["P(KP481|income_range)"]=(x["KP481"]/x["All"]*100).round(2)
x["P(KP781|income_range)"]=(x["KP781"]/x["All"]*100).round(2)
x=x.drop([4])
x=x.drop("All",axis=1)
x
```

## Conditional probability of product purchased by certain age group

```python
y=pd.crosstab(df_new.Age,df_new.Product,margins=True).reset_index()
y
```

```python
y["P(KP281|age_group)"]=(y["KP281"]/y["All"]*100).round(2)
y["P(KP481|age_group)"]=(y["KP481"]/y["All"]*100).round(2)
y["P(KP781|age_group)"]=(y["KP781"]/y["All"]*100).round(2)
y=y.drop([4])
y=y.drop("All",axis=1)
y
```
